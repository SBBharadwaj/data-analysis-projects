Transactions dataset

```python
np.random.seed(42)
n=20000
transaction_id = productid
customer_id= custid
product_id=productid
quantity=np.random.randint(1,10,n)
transaction_date=pd.date_range(start='01-01-2024',periods=n,freq='h')
payment_method=np.random.choice(['Cash','Credit Card','Debit Card','UPI'],n)
discount=np.random.uniform(0.1,0.5,n).round(2)
total_amount=price*quantity*(1-discount).round(2)
store_location=city


transactions = pd.DataFrame({'Transaction_id':transaction_id,
                          'Customer_Id':customer_id,
                          'Product_ID':product_id,
                          'Quantity':quantity,
                          'Transaction_Date':transaction_date,
                          'Payment_Method':payment_method,
                          'Discount':discount,
                          'Total_Amount':total_amount,
                          'Store_Location':store_location})
```

```python
#50. Build a dashboard including:
#    - Total Revenue
#    - Total Customers
#    - Top Products
#    - Monthly Trends
#    - Customer Segments

transactions['Transaction_Date'] = pd.to_datetime(transactions['Transaction_Date'])

# KPI values
total_revenue = transactions['Total_Amount'].sum()
total_customers = customers['customer_id'].nunique()

#Monthly trend
monthly_sales = transactions.groupby(
    transactions['Transaction_Date'].dt.month)['Total_Amount'].sum()

#Top products
top_products = transactions['Product_ID'].value_counts().head(5)

#Customer Segmentation
spending = transactions.groupby('Customer_Id')['Total_Amount'].sum()

def segment(x):
    if x > 50000:
        return 'High'
    elif x > 20000:
        return 'Medium'
    else:
        return 'Low'

segments = spending.apply(segment).value_counts()

#Dashboard layout
plt.figure(figsize=(10,10))

#Monthly Trends
plt.subplot(3,2,1)
plt.plot(monthly_sales.index, monthly_sales.values)
plt.title("Monthly Sales Trend")

#Top Products
plt.subplot(3,2,2)
sns.barplot(x=top_products.index, y=top_products.values)
plt.title("Top Products")

#Customer Segments
plt.subplot(3,2,3)
sns.barplot(x=segments.index, y=segments.values)
plt.title("Customer Segmentation")

#KPIs
plt.subplot(3,2,4)
plt.text(0.1, 0.6, f"Total Revenue:\n{total_revenue}", fontsize=12)
plt.text(0.1, 0.3, f"Total Customers:\n{total_customers}", fontsize=12)
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Sales trend over time
merged['Transaction_Date'] = pd.to_datetime(merged['Transaction_Date'])
merged['YearMonth'] = merged['Transaction_Date'].dt.strftime('%Y-%b')
sales = merged.groupby('YearMonth')['Total_Amount'].sum()
# Ensure chronological order for plotting
sales = sales.reindex(pd.to_datetime(sales.index).sort_values().strftime('%Y-%b'))
sales
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=sales.index, y=sales.values)
plt.title('Sales trend over time')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45, ha='right')
plt.grid(True)

# Find month with highest sales
max_sales_month = sales.idxmax()
max_sales_value = sales.max()
max_sales_month_loc = sales.index.get_loc(max_sales_month)
plt.annotate(f'High: ${max_sales_value:,.0f}', xy=(max_sales_month_loc, max_sales_value), xytext=(max_sales_month_loc + 1.2, max_sales_value + sales.max()*0.15),
             arrowprops=dict(facecolor='red', shrink=0.2), fontsize=10, color='red')

# Find month with lowest sales
min_sales_month = sales.idxmin()
min_sales_value = sales.min()
min_sales_month_loc = sales.index.get_loc(min_sales_month)
plt.annotate(f'Low: ${min_sales_value:,.0f}', xy=(min_sales_month_loc, min_sales_value), xytext=(min_sales_month_loc + 1.2, min_sales_value - sales.max()*0.12),
             arrowprops=dict(facecolor='green', shrink=0.2), fontsize=10, color='green')

plt.tight_layout()
plt.show()
```
