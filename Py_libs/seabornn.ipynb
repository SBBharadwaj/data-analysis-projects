```python
df=sns.load_dataset('tips')
```

```python
df
```

```python
sns.set(style='darkgrid',context='paper')
sns.countplot(data=df,x='day',hue='sex',order=['Fri','Thur'],palette='Reds')
```

```python
sns.set(style='darkgrid',context='paper')
sns.barplot(data=df,x='day',y='total_bill',estimator='max',errorbar='sd',hue='sex',palette='Blues')
```

```python
sns.set(style='darkgrid',context='notebook')
sns.swarmplot(data=df,x='day',y='total_bill',hue='smoker',palette='Greens',dodge=False)
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',style='smoker',hue='sex',palette='viridis',size='sex',sizes=(40,50))
plt.legend(bbox_to_anchor=(1,1))
```
